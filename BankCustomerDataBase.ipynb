**Creating the database**

```sql
--Create a new Database
CREATE DATABASE BankCustomer;
GO

-- Start working with the database by calling it
USE BankCustomer;
GO

-- Creating the various tables used in the database
CREATE TABLE [Customer](
    [CustomerId] INT IDENTITY(1,1) NOT NULL PRIMARY KEY 
    ,[CustomerFirstName] NVARCHAR(300) NOT NULL
    ,[CustomerLastName] NVARCHAR(300) NOT NULL
    ,[CustomerBirthDate] NVARCHAR(8) NULL
);
--Using ISO codes to identify the Adress country
CREATE TABLE [CustomerAdress](
    [CustomerId] INT NOT NULL PRIMARY KEY 
    ,[PrimaryAdressCountry] NVARCHAR(10) NOT NULL FOREIGN KEY REFERENCES Customer(CustomerId)
    ,[SecondaryAdressCountry] NVARCHAR(10) NULL FOREIGN KEY REFERENCES Customer(CustomerId)
);

--Using ISO codes to identify the Tax country
CREATE TABLE [CustomerTaxCountry](
    [CustomerId] INT NOT NULL PRIMARY KEY 
    ,[PrimaryTaxCountry] NVARCHAR(10) NOT NULL FOREIGN KEY REFERENCES CustomerAdressScore(AdressCountry)
    ,[SecondaryTaxCountry] NVARCHAR(10) NULL FOREIGN KEY REFERENCES CustomerAdressScore(AdressCountry)
);

--Using ISO codes to identify the Citizenship country
CREATE TABLE [CustomerCitizenship](
    [CustomerId] INT NOT NULL PRIMARY KEY 
    ,[PrimaryCitizenship] NVARCHAR(10) NOT NULL FOREIGN KEY REFERENCES Customer(CustomerId)
    ,[SecondaryCitizenship] NVARCHAR(10) NULL FOREIGN KEY REFERENCES Customer(CustomerId)
);

--Using ISO codes to identify the Adress country score in order to calculate the customers risk score
CREATE TABLE [CustomerAdressScore](
    [AdressCountry] NVARCHAR(10) NOT NULL PRIMARY KEY 
    ,[AdressCountryScore] INT NOT NULL 
);

--Using ISO codes to identify the Tax country score in order to calculate the customers risk score
CREATE TABLE [CustomerTaxCountryScore](
    [TaxCountry] NVARCHAR(10) NOT NULL PRIMARY KEY 
    ,[TaxCountryScore] INT NOT NULL 
);

--Using ISO codes to identify the Citizenship score in order to calculate the customers risk score
CREATE TABLE [CustomerCitizenshipScore](
    [Citizenship] NVARCHAR(10) NOT NULL PRIMARY KEY 
    ,[CitizenshipScore] INT NOT NULL 
);

```

**Bank Customer database overview**

**Creating the database**

```sql
--Create a new Database
CREATE DATABASE BankCustomer;
GO

-- Start working with the database by calling it
USE BankCustomer;
GO

-- Creating the various tables used in the database
CREATE TABLE [Customer](
    [CustId] INT IDENTITY(1,1) NOT NULL PRIMARY KEY 
    ,[CustomerFirstName] NVARCHAR(300) NOT NULL
    ,[CustomerLastName] NVARCHAR(300) NOT NULL
    ,[CustomerBirthDate] NVARCHAR(8) NULL
);

--Using ISO codes to identify the Adress country score in order to calculate the customers risk score
CREATE TABLE [CustomerAdressScore](
    [AdressCountry] NVARCHAR(2) NOT NULL PRIMARY KEY 
    ,[AdressCountryScore] INT NOT NULL 
);

--Using ISO codes to identify the Tax country score in order to calculate the customers risk score
CREATE TABLE [CustomerTaxCountryScore](
    [TaxCountry] NVARCHAR(2) NOT NULL PRIMARY KEY 
    ,[TaxCountryScore] INT NOT NULL 
);

--Using ISO codes to identify the Citizenship score in order to calculate the customers risk score
CREATE TABLE [CustomerCitizenshipScore](
    [Citizenship] NVARCHAR(2) NOT NULL PRIMARY KEY 
    ,[CitizenshipScore] INT NOT NULL 
);

--In order to get the actual name for the ISO code
CREATE TABLE [CountryMap](
    [ISOCode] NVARCHAR(2) NOT NULL PRIMARY KEY 
    ,[ISOName] NVARCHAR(100) NOT NULL 
);

--Using ISO codes to identify the Adress country
CREATE TABLE [CustomerAdress](
    [CustId] INT NOT NULL PRIMARY KEY 
    ,[PrimaryAdressCountry] NVARCHAR(2) NOT NULL 
    ,[SecondaryAdressCountry] NVARCHAR(2) NULL 
);

--Using ISO codes to identify the Tax country
CREATE TABLE [CustomerTaxCountry](
    [CustId] INT NOT NULL PRIMARY KEY 
    ,[PrimaryTaxCountry] NVARCHAR(2) NOT NULL 
    ,[SecondaryTaxCountry] NVARCHAR(2) NULL 
);

--Using ISO codes to identify the Citizenship country
CREATE TABLE [CustomerCitizenship](
    [CustId] INT NOT NULL PRIMARY KEY 
    ,[PrimaryCitizenship] NVARCHAR(2) NOT NULL 
    ,[SecondaryCitizenship] NVARCHAR(2) NULL 
);



```

**Creating a Stored procedure to calculate the risk score and return the final risk rating**

```sql
IF EXISTS (
SELECT *
    FROM INFORMATION_SCHEMA.ROUTINES
WHERE SPECIFIC_SCHEMA = N'dbo'
    AND SPECIFIC_NAME = N'STP_RiskRating'
    AND ROUTINE_TYPE = N'PROCEDURE'
)
DROP PROCEDURE dbo.STP_RiskRating
GO

CREATE PROCEDURE dbo.STP_RiskRating
     @CustId INT = 1
    ,@FinalRiskRating NVARCHAR(10) = FinalRiskRating
AS
BEGIN
 
SELECT CustId 
,FinalRiskRating
FROM
(
SELECT CustId
,CASE WHEN  FinalRiskScore<=5
    THEN 'LOW' 
        WHEN FinalRiskScore BETWEEN 5 AND 10 
            THEN 'MEDIUM'
                WHEN FinalRiskScore>10 
                    THEN 'HIGH' END 
            AS FinalRiskRating 
FROM
    (
SELECT adress.*
, TaxCountries
, FinalTaxCountryScore
, Citizenships
, FinalCitizenshipscore
, (Finaladresscountryscore+FinalTaxCountryScore+FinalCitizenshipscore) AS FinalRiskScore
    FROM
        (
SELECT pasc.custid
, concat(pasc.primaryadresscountry,',',sasc.secondaryadresscountry) as AdressCountries
, CASE WHEN ISNULL(primaryadresscountryscore,0)>=ISNULL(secondaryadresscountryscore,0)  THEN primaryadresscountryscore
    ELSE secondaryadresscountryscore END AS FinalAdressCountryScore
        FROM
            ( 
SELECT custid
, primaryadresscountry
, adresscountryscore AS primaryadresscountryscore
            FROM BankCustomer.dbo.CustomerAdress AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerAdressScore AS CAS
                ON CA.primaryadressCountry=CAS.adresscountry
) AS PASC
            LEFT JOIN
            (
SELECT custid
, secondaryadresscountry
, adresscountryscore AS secondaryadresscountryscore
            FROM BankCustomer.dbo.CustomerAdress AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerAdressScore AS CAS
                ON CA.secondaryadressCountry=CAS.adresscountry
) AS SASC
            ON PASC.custId=SASC.custId
) AS Adress

        LEFT JOIN
        (
SELECT ptsc.custid
, concat(ptsc.primaryTaxcountry,',',stsc.secondaryTaxcountry) AS TaxCountries
, CASE WHEN ISNULL(primaryTaxcountryscore,0)>=ISNULL(secondaryTaxcountryscore,0)  THEN primaryTaxcountryscore
    ELSE secondaryTaxcountryscore END AS FinalTaxCountryScore
        FROM
            ( 
SELECT custid
, primaryTaxcountry
, TaxcountryScore AS primaryTaxcountryscore
            FROM BankCustomer.dbo.CustomerTaxCountry AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerTaxCountryScore AS CAS
                ON CA.primaryTaxCountry=CAS.Taxcountry
) AS PTSC
            LEFT JOIN
            (
SELECT custid
, secondaryTaxcountry
, TaxcountryScore AS secondaryTaxcountryscore
            FROM BankCustomer.dbo.CustomerTaxCountry AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerTaxCountryScore AS CAS
                ON CA.secondaryTaxCountry=CAS.Taxcountry
) AS STSC
            ON PTSC.custId=STSC.custId
) AS Tax
        ON Adress.CustId=Tax.CustId

        LEFT JOIN
        (
SELECT pcsc.custid
, concat(pcsc.primaryCitizenship,',',scsc.secondaryCitizenship) AS Citizenships
, CASE WHEN ISNULL(primaryCitizenshipscore,0)>=ISNULL(secondaryCitizenshipscore,0)  THEN primaryCitizenshipscore
    ELSE secondaryCitizenshipscore END AS FinalCitizenshipScore
        FROM
            ( 
SELECT custid
, primaryCitizenship
, CitizenshipScore AS primaryCitizenshipscore
            FROM BankCustomer.dbo.CustomerCitizenship AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerCitizenshipScore AS CAS
                ON CA.PrimaryCitizenship=CAS.Citizenship
) AS PCSC
            LEFT JOIN
            (
SELECT custid
, secondaryCitizenship
, CitizenshipScore AS secondaryCitizenshipscore
            FROM BankCustomer.dbo.CustomerCitizenship AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerCitizenshipScore AS CAS
                ON CA.SecondaryCitizenship=CAS.Citizenship
) AS SCSC
            ON PCSC.custId=SCSC.custId
) AS Citi
        on Adress.CustId=Citi.CustId
) AS FINAL
) AS F
WHERE FinalRiskRating = @FinalRiskRating 
    
END
GO

/*Second Stored procedure to return information on customernumber*/
IF EXISTS (
SELECT *
    FROM INFORMATION_SCHEMA.ROUTINES
WHERE SPECIFIC_SCHEMA = N'dbo'
    AND SPECIFIC_NAME = N'STP_CustId'
    AND ROUTINE_TYPE = N'PROCEDURE'
)
DROP PROCEDURE dbo.STP_CustId
GO

CREATE PROCEDURE dbo.STP_CustId
     @CustId INT = 1
    ,@FinalRiskRating NVARCHAR(10) = FinalRiskRating
AS
BEGIN
 
SELECT CustId 
,FinalRiskRating
FROM
(
SELECT CustId
,CASE WHEN  FinalRiskScore<=5
    THEN 'LOW' 
        WHEN FinalRiskScore BETWEEN 5 AND 10 
            THEN 'MEDIUM'
                WHEN FinalRiskScore>10 
                    THEN 'HIGH' END 
            AS FinalRiskRating 
FROM
    (
SELECT adress.*
, TaxCountries
, FinalTaxCountryScore
, Citizenships
, FinalCitizenshipscore
, (Finaladresscountryscore+FinalTaxCountryScore+FinalCitizenshipscore) AS FinalRiskScore
    FROM
        (
SELECT pasc.custid
, concat(pasc.primaryadresscountry,',',sasc.secondaryadresscountry) as AdressCountries
, CASE WHEN ISNULL(primaryadresscountryscore,0)>=ISNULL(secondaryadresscountryscore,0)  THEN primaryadresscountryscore
    ELSE secondaryadresscountryscore END AS FinalAdressCountryScore
        FROM
            ( 
SELECT custid
, primaryadresscountry
, adresscountryscore AS primaryadresscountryscore
            FROM BankCustomer.dbo.CustomerAdress AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerAdressScore AS CAS
                ON CA.primaryadressCountry=CAS.adresscountry
) AS PASC
            LEFT JOIN
            (
SELECT custid
, secondaryadresscountry
, adresscountryscore AS secondaryadresscountryscore
            FROM BankCustomer.dbo.CustomerAdress AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerAdressScore AS CAS
                ON CA.secondaryadressCountry=CAS.adresscountry
) AS SASC
            ON PASC.custId=SASC.custId
) AS Adress

        LEFT JOIN
        (
SELECT ptsc.custid
, concat(ptsc.primaryTaxcountry,',',stsc.secondaryTaxcountry) AS TaxCountries
, CASE WHEN ISNULL(primaryTaxcountryscore,0)>=ISNULL(secondaryTaxcountryscore,0)  THEN primaryTaxcountryscore
    ELSE secondaryTaxcountryscore END AS FinalTaxCountryScore
        FROM
            ( 
SELECT custid
, primaryTaxcountry
, TaxcountryScore AS primaryTaxcountryscore
            FROM BankCustomer.dbo.CustomerTaxCountry AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerTaxCountryScore AS CAS
                ON CA.primaryTaxCountry=CAS.Taxcountry
) AS PTSC
            LEFT JOIN
            (
SELECT custid
, secondaryTaxcountry
, TaxcountryScore AS secondaryTaxcountryscore
            FROM BankCustomer.dbo.CustomerTaxCountry AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerTaxCountryScore AS CAS
                ON CA.secondaryTaxCountry=CAS.Taxcountry
) AS STSC
            ON PTSC.custId=STSC.custId
) AS Tax
        ON Adress.CustId=Tax.CustId

        LEFT JOIN
        (
SELECT pcsc.custid
, concat(pcsc.primaryCitizenship,',',scsc.secondaryCitizenship) AS Citizenships
, CASE WHEN ISNULL(primaryCitizenshipscore,0)>=ISNULL(secondaryCitizenshipscore,0)  THEN primaryCitizenshipscore
    ELSE secondaryCitizenshipscore END AS FinalCitizenshipScore
        FROM
            ( 
SELECT custid
, primaryCitizenship
, CitizenshipScore AS primaryCitizenshipscore
            FROM BankCustomer.dbo.CustomerCitizenship AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerCitizenshipScore AS CAS
                ON CA.PrimaryCitizenship=CAS.Citizenship
) AS PCSC
            LEFT JOIN
            (
SELECT custid
, secondaryCitizenship
, CitizenshipScore AS secondaryCitizenshipscore
            FROM BankCustomer.dbo.CustomerCitizenship AS CA
                INNER JOIN
                BankCustomer.dbo.CustomerCitizenshipScore AS CAS
                ON CA.SecondaryCitizenship=CAS.Citizenship
) AS SCSC
            ON PCSC.custId=SCSC.custId
) AS Citi
        on Adress.CustId=Citi.CustId
) AS FINAL
) AS F
WHERE CustId = @CustId 
    
END
GO

-- Customer Overview SP

IF EXISTS (
SELECT *
    FROM INFORMATION_SCHEMA.ROUTINES
WHERE SPECIFIC_SCHEMA = N'dbo'
    AND SPECIFIC_NAME = N'STP_CustOverview'
    AND ROUTINE_TYPE = N'PROCEDURE'
)
DROP PROCEDURE dbo.STP_CustOverview
GO

CREATE PROCEDURE dbo.STP_CustOverview
    @CustId int

AS
BEGIN
    SELECT C.*
, ISNULL(PrimaryCitizenship,'NOT APPLICABLE') AS PrimaryCitizenship
, ISNULL(SecondaryCitizenship,'NOT APPLICABLE') AS SecondaryCitizenship
, ISNULL(PrimaryAdressCountry,'NOT APPLICABLE') AS PrimaryAdressCountry
, ISNULL(SecondaryAdressCountry,'NOT APPLICABLE') AS SecondaryAdressCountry
, ISNULL(PrimaryTaxCountry,'NOT APPLICABLE') AS PrimaryTaxCountry
, ISNULL(SecondaryTaxCountry,'NOT APPLICABLE') AS SecondaryTaxCountry
    FROM BankCustomer.dbo.Customer as c
        left JOIN
        (
SELECT CustId 
, CONCAT(tc.PrimaryTaxCountry,'-',cm.ISOName) as PrimaryTaxCountry
, CONCAT(tc.SecondaryTaxCountry,'-',cm1.ISOName) as SecondaryTaxCountry
        FROM
            BankCustomer.dbo.CustomerTaxCountry as tc
            INNER JOIN
            BankCustomer.dbo.CountryMap AS CM
            ON tc.PrimaryTaxCountry=cm.ISOCode
            INNER JOIN
            BankCustomer.dbo.CountryMap AS CM1
            ON tc.SecondaryTaxCountry=cm1.ISOCode
) AS TC
        ON c.CustId=tc.CustId
        LEFT JOIN
        (
SELECT CustId 
, CONCAT(CC.PrimaryCitizenship,'-',cm.ISOName) as PrimaryCitizenship
, CONCAT(CC.SecondaryCitizenship,'-',cm1.ISOName) as SecondaryCitizenship
        FROM
            BankCustomer.dbo.CustomerCitizenship as CC
            INNER JOIN
            BankCustomer.dbo.CountryMap AS CM
            ON CC.PrimaryCitizenship=cm.ISOCode
            INNER JOIN
            BankCustomer.dbo.CountryMap AS CM1
            ON CC.SecondaryCitizenship=cm1.ISOCode
) AS CC
        ON C.CustId=CC.CustId
        LEFT JOIN
        (
SELECT CustId 
, CONCAT(AC.PrimaryAdressCountry,'-',cm.ISOName) as PrimaryAdressCountry
, CONCAT(AC.SecondaryAdressCountry,'-',cm1.ISOName) as SecondaryAdressCountry
        FROM
            BankCustomer.dbo.CustomerAdress as AC
            INNER JOIN
            BankCustomer.dbo.CountryMap AS CM
            ON AC.PrimaryAdressCountry=cm.ISOCode
            INNER JOIN
            BankCustomer.dbo.CountryMap AS CM1
            ON AC.SecondaryAdressCountry=cm1.ISOCode
) AS AC
        ON C.CustId=AC.CustId
    Where c.CustId=@CustId
END
GO

```
